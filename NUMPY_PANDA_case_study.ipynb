#HANDILING MISSING DATA

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("auto-mpg (2).csv")

# Step 1: Check missing values (initial)
print(df.isnull().sum())

# Step 2: Convert 'horsepower' to numeric (IMPORTANT)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Step 3: Check missing values again (after conversion)
print(df.isnull().sum())

# Step 4: Fill missing values with median
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)

# Step 5: Final check
print(df.isnull().sum())
```

#DATA TRANSFORMATION

```python
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
print(df['power_to_weight_ratio'])
```

#GROUP BY

#APPLY FUCTION

```python
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)
df['performance_score']
```

#VISUALISATION PREPARATION

```python
import pandas as pd

# Load dataset
df = pd.read_csv("auto-mpg (2).csv")

# Clean column names (fixes model_year issue)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Fix horsepower column (handles '?' values)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)

# -------------------------------
# Visualization Preparation
# -------------------------------
summary = df.groupby('model_year')[['mpg', 'weight', 'horsepower']].mean()

# Display result
print(summary)
```

#EXPORTING DATA

```python
high_mpg = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg.to_csv("high_mpg_cars.csv", index=False)
```

#FINDING ANOMALIES

```python
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['mpg'] < Q1 - 1.5*IQR) | (df['mpg'] > Q3 + 1.5*IQR)]

result = outliers[['car_name', 'mpg', 'model_year']]
print(result)
```

#BROADCASTING

```python
import numpy as np

hp = pd.to_numeric(df['horsepower'], errors='coerce').values
hp = np.where(np.isnan(hp), np.nanmean(hp), hp)

hp_new = hp * 1.10

print(hp_new[:10])
```
